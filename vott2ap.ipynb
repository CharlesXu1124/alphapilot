Convert VoTT custom vision format to Alphapilot label

```python
import pandas as pd
import json
```

```python
df = pd.read_json('/Users/danielnbarbosa/.fastai/data/training_superclean_labels.json')
```

```python
df.head()
```

```python
df_new = pd.DataFrame()
```

```python
for key in df.keys():
    points = df[key][1]['points']
    coords = [points[0]['x'], points[0]['y'],
              points[1]['x'], points[1]['y'], 
              points[2]['x'], points[2]['y'],
              points[3]['x'], points[3]['y']]
    df_new[key] = [[coords]]
    
```

```python
df_new.head()
```

```python
# write out clean json file with all the community labels

# convert dataframe to JSON string and strip out outer '[]'
json_str = df_new.to_json(orient='records')[1:-1]

# write to file
f = open('/Users/danielnbarbosa/.fastai/data/training_superclean_labels_converted.json', 'w')
f.write(json_str)
f.close()
```
